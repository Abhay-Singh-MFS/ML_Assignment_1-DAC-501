# Importing necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options to show all columns
pd.set_option('display.max_columns', None)

# Set display options to show all rows
pd.set_option('display.max_rows', None)
```

# Data Loading and Exploration

```python
# Load the dataset
df = pd.read_csv('housing_prices.csv')
```

```python
# Display the first few rows
print(df.head())
```

```python
# Check the structure of the dataset
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

### Choosing the variables for the analysis

```python
# Numerical variables
numerical_cols = ['LotArea', 'YearBuilt', 'GrLivArea', 'SalePrice', 'OverallQual']

# Categorical variables
categorical_cols = ['MSZoning', 'Neighborhood', 'HouseStyle', 'SaleCondition']
```

# Data Cleaning and further exploration

```python
# Specify the columns to check for null values
columns_to_check = ['LotArea', 'YearBuilt', 'GrLivArea', 'SalePrice', 
                     'OverallQual', 'MSZoning', 'Neighborhood', 
                     'HouseStyle', 'SaleCondition']

# Drop rows where any of the specified columns have null values
df_cleaned = df.dropna(subset=columns_to_check)

# Verify the missing values are handled
print(df_cleaned.isnull().sum())
```

```python
df_cleaned.info()
```

```python
# Check unique values for categorical columns
for col in categorical_cols:
    print(f'{col}: {df_cleaned[col].nunique()} unique values')
```

```python
# Explore potential outliers using summary statistics
print(df_cleaned[numerical_cols].describe())
```

### Outlier Detection

```python
# Boxplot to detect outliers in numerical columns
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df_cleaned[col])
    plt.title(f'Boxplot of {col} (Outlier Detection)')
    plt.show()
```

```python
# Detect outliers using IQR (Interquartile Range) method
Q1 = df_cleaned[numerical_cols].quantile(0.25)
Q3 = df_cleaned[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Filtering out the outliers
df_no_outliers = df_cleaned[~((df_cleaned[numerical_cols] < (Q1 - 1.5 * IQR)) |(df_cleaned[numerical_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f'Rows before removing outliers: {len(df_cleaned)}')
print(f'Rows after removing outliers: {len(df_no_outliers)}')
```

```python
df_cleaned = df_no_outliers.copy()
```

# Univariate Analysis

### Feature Distribution and Skewness

```python
# Calculate skewness of numerical columns
print(df_cleaned[numerical_cols].skew())
```

```python
# Plot histograms with KDE (Kernel Density Estimate)
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(df_cleaned[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Count plot for each categorical column
for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df_cleaned, x=col)
    plt.title(f'Count Plot for {col}')
    plt.xticks(rotation=45)
    plt.show()
```

# Multivariate Analysis

```python
# Scatter plot between GrLivArea and SalePrice
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='GrLivArea', y='SalePrice')
plt.title('Scatter Plot of GrLivArea vs SalePrice')
plt.show()

# Scatter plot of SalePrice vs LotArea
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LotArea', y='SalePrice', data=df_cleaned)
plt.title('SalePrice vs LotArea')
plt.xlabel('Lot Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.show()
```

### Analyzing SalePrice with Different Categories

```python
# Boxplot of SalePrice by Neighborhood
plt.figure(figsize=(10, 6))
sns.boxplot(x='Neighborhood', y='SalePrice', data=df_cleaned)
plt.title('SalePrice by Neighborhood')
plt.xticks(rotation=45)
plt.show()

# Boxplot of SalePrice by Overall Quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='OverallQual', y='SalePrice', data=df_cleaned)
plt.title('SalePrice by Overall Quality')
plt.xticks(rotation=45)
plt.show()

# Boxplot of SalePrice by HouseStyle
plt.figure(figsize=(10, 6))
sns.boxplot(x='HouseStyle', y='SalePrice', data=df_cleaned)
plt.title('SalePrice by HouseStyle')
plt.xticks(rotation=45)
plt.show()

# Boxplot of SalePrice by SaleCondition
plt.figure(figsize=(10, 6))
sns.boxplot(x='SaleCondition', y='SalePrice', data=df_cleaned)
plt.title('SalePrice by SaleCondition')
plt.xticks(rotation=45)
plt.show()

# Barplot of average SalePrice by MSZoning
plt.figure(figsize=(10, 6))
sns.barplot(x='MSZoning', y='SalePrice', data=df_cleaned, estimator=sum)
plt.title('Sum of SalePrice by MSZoning')
plt.show()
```

### Feature Engineering (Creating New Features)

```python
# Create a new feature - AgeOfHouse
df_cleaned['AgeOfHouse'] = 2024 - df_cleaned['YearBuilt']

# Visualize the relationship between AgeOfHouse and SalePrice
plt.figure(figsize=(8, 6))
sns.scatterplot(x='AgeOfHouse', y='SalePrice', data=df_cleaned)
plt.title('SalePrice vs AgeOfHouse')
plt.show()
```

### Correlation Analysis

```python
# Correlation heatmap including new feature
extended_numerical_cols = numerical_cols + ['AgeOfHouse']
plt.figure(figsize=(10, 8))
sns.heatmap(df_cleaned[extended_numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
